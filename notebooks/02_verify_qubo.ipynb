```python
import os
import sys
import pickle
sys.path.append(os.path.abspath('..'))

import numpy as np
from src.qubo import build_qubo
from src.ising import qubo_to_ising
```

```python
with open('results/raw/instance_data.pkl', 'rb') as f:
    data_n1 = pickle.load(f)

scenarios = data_n1['scenarios']
alpha = data_n1['alpha']
print("Data instance berhasil di-load!")
print(f"alpha: {alpha} | scenarios = {scenarios}" )
```

```python
m = 30              # Jumlah aksi (qubits)
costs = np.ones(m)  # Asumsi cost seragam untuk eksperimen MVD
budget = 5.0        # Batas anggaran
eta = 10.0          # Multiplier penalti anggaran
```

```python
q, Q, const = build_qubo(scenarios, alpha, costs, budget, eta)
K, g, c_ising = qubo_to_ising(q, Q, const)

print("QUBO Linear Terms (q):", q.shape)
print("QUBO Quadratic Terms (Q):", Q.shape)
print("Konstanta Ising:", c_ising)
```

```python
with open('results/raw/qubo_ising_data.pkl', 'wb') as f:
    pickle.dump({
        'q': q, 'Q': Q, 'const': const,
        'K': K, 'g': g, 'c_ising': c_ising
    }, f)
print("Data QUBO dan Ising berhasil disimpan ke pickle!")
```
